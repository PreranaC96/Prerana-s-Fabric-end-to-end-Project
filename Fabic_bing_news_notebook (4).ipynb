Read Json file from lake database as a Dataframe

```python
df = spark.read.option("multiline", "true").json("Files/fabric_bing_news.json")
# df now is a Spark DataFrame containing JSON data from "Files/fabric_bing_news.json".
display(df)
```

```python
df = df.select("value")
```

```python
display(df)
```

Explode the Json column

```python
from pyspark.sql.functions import explode
df_exploded_bing = df.select(explode(df["value"]).alias("json_bing"))
```

```python
display(df_exploded_bing)
```

Converting exploded Json DF to Single Json string list

```python
json_bing_list = df_exploded_bing.toJSON().collect()
#json_list
```

```python
print(json_bing_list[20])
```

```python
import json

news_json = json.loads(json_bing_list[20])
```

```python
print (news_json)
```

```python
print (news_json['json_bing']['name'])
print (news_json['json_bing']['description'])
print (news_json['json_bing']['url'])
print (news_json['json_bing']["image"]["thumbnail"]["contentUrl"])
print (news_json['json_bing']['provider'][0]['name'])
print (news_json['json_bing']['datePublished'])
```

Processing the Json property list

```python
title =[]
description =[]
image =[]
provider =[]
url =[]
datePublished =[]


#process each Json object in the list

for json_str in json_bing_list:
    try:
        #parse json string into a dictionary
        article = json.loads(json_str)

        #extract information from the dictionary
        title.append(article['json_bing']['name'])
        description.append(article['json_bing']['description'])
        url.append(article['json_bing']['url'])
        ##image.append(article['json_bing']["image"]["thumbnail"]["contentUrl"])
        provider.append(article['json_bing']['provider'][0]['name'])
        datePublished.append(article['json_bing']['datePublished'])

    except Exception as e:
        print(f"Error processing JSON object:{e}")    
```

Converting list into Data frame

```python
from pyspark.sql.types import StructType, StructField, StringType

#Combine the list
data = list(zip(title,description,url,provider,datePublished))

#Define schema
schema = StructType([
    StructField('title',StringType(),True),
    StructField('description',StringType(),True),
    StructField('url',StringType(),True),
    StructField('provider',StringType(),True),
    StructField('datePublished',StringType(),True)
])

#Create DataFrame
df_bing_cleaned = spark.createDataFrame(data, schema=schema)
```

```python
display(df_bing_cleaned)
```

Processing Date column

```python
from pyspark.sql.functions import to_date, date_format

df_cleaned_final = df_bing_cleaned.withColumn("datePublished",date_format(to_date("datePublished"),"dd-MMM-yyyy"))
```

```python
display(df_cleaned_final)
```

```python
df_cleaned_final.write.format("delta").mode('overwrite').saveAsTable("fabric_bing_lake_db.tbl_bing_latest_news")
```
